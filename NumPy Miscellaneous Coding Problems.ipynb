# Q1) Two Sum Problem.

- Find two indices whose values add up to a given Target.

```python
def two_sum(nums, target):
    num_map = {}  # Stores number and its index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []

nums = [2, 7, 11, 15]
target = 9
print("Indices of Two Sum:", two_sum(nums, target))
```

# Q2) Maximum Subarray Sum (Kadane's Algorithm).

-  Find the contiguous subarray in 1-D NumPy Array with the Largest Sum.

```python
def max_subarray_sum(nums):
    max_sum = curr_sum = nums[0]
    for num in nums[1:]:
        curr_sum = max(num, curr_sum + num)
        max_sum = max(max_sum, curr_sum)
    return max_sum

nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print("Maximum Subarray Sum:", max_subarray_sum(nums))
```

# Q3) Product of Array Except Self.

-  Find an array where each index is the product of all numbers except itself.

```python
def product_except_self(nums):
    n = len(nums)
    result = [1] * n
    left_product = 1
    
    for i in range(n):
        result[i] = left_product
        left_product *= nums[i]

    right_product = 1
    for i in range(n-1, -1, -1):
        result[i] *= right_product
        right_product *= nums[i]

    return result

nums = [1, 2, 3, 4]
print("Product Except Self:", product_except_self(nums))
```

# Q4) Find the First Missing Positive Integer.

-  Find the smallest missing positive integer.

```python
def first_missing_positive(nums):
    n = len(nums)
    for i in range(n):
        while 1 <= nums[i] <= n and nums[i] != nums[nums[i] - 1]:
            nums[nums[i] - 1], nums[i] = nums[i], nums[nums[i] - 1]
    
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1
    return n + 1

nums = [3, 4, -1, 1]
print("First Missing Positive:", first_missing_positive(nums))
```
