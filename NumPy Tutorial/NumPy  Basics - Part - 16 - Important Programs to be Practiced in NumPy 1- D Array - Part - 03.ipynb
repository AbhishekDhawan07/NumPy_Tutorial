# Q31) Find the Index of the Maximum and Minimum Element in NumPy Array



```python
import numpy as np

arr = np.array([10, 20, 30, 5, 25, 50])
max_index = np.argmax(arr)
min_index = np.argmin(arr)

print("Index of Maximum Element:", max_index)
print("Index of Minimum Element:", min_index)
```

# Q35) Find the Percentage Contribution of Each Element in an NumPy Array

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
percentage_contribution = (arr / np.sum(arr)) * 100

print("Percentage Contribution:", percentage_contribution)
```

# Q36) Count Elements Within a Given Range

```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25, 30, 35])
lower_bound = 10
upper_bound = 30

count_in_range = np.count_nonzero((arr >= lower_bound) & (arr <= upper_bound))

print(f"Count of Elements in Range [{lower_bound}, {upper_bound}]:", count_in_range)
```

# Q38) Normalize an Array Using Z-Score Normalization

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
mean = np.mean(arr)
std_dev = np.std(arr)
z_score_normalized = (arr - mean) / std_dev  # Compute Z-score normalization

print("Z-Score Normalized Array:", z_score_normalized)
```

# Q39) Find the Angles (in Degrees) for Given Sine Values

```python
import numpy as np

sine_values = np.array([0, 0.5, 1])
angles = np.degrees(np.arcsin(sine_values))  # Compute inverse sine (arcsin) and convert to degrees

print("Angles in Degrees:", angles)
```

# Q40) Detect Outliers in a 1-D Array Using the IQR Method

```python
import numpy as np

arr = np.array([10, 12, 15, 18, 22, 100, 105])  # 100 and 105 are potential outliers

Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = arr[(arr < lower_bound) | (arr > upper_bound)]

print("Outliers:", outliers)
```
