# Q16) Check if Two 1- D Arrays are Disjoint(No Common Elements) 

```python
import numpy as np

arr1 = np.array([1, 3, 5, 7])
arr2 = np.array([2, 4, 6, 8])

is_disjoint = np.intersect1d(arr1, arr2).size == 0

print("Are the arrays disjoint?", is_disjoint)
```

# Q19) Normalize a 1-D Array

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
normalized_arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))

print("Normalized Array:", normalized_arr)
```

# Q22) Find the Index of the First Occurrence of an Element

```python
import numpy as np

arr = np.array([4, 7, 9, 2, 7, 3, 7])
element = 7

index = np.where(arr == element)[0][0]  # First occurrence index

print(f"First occurrence of {element} is at index:", index)
```

# Q24) Find Elements that Appear More than Once(Non - Unique Elements) in NumPy Array.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 2, 3, 5, 6, 1])
unique_elements, counts = np.unique(arr, return_counts=True)
duplicates = unique_elements[counts > 1]

print("Duplicate Elements:", duplicates)
```

# Q25) Convert an NumPy Array to a Binary Representation

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
binary_representation = [bin(x)[2:] for x in arr]  # Convert each element to binary

print("Binary Representation:", binary_representation)
```

# Q26) Find the Greatest Common Divisor (GCD) of an NumPy Array

```python
import numpy as np

arr = np.array([8, 16, 32])
gcd_value = np.gcd.reduce(arr)

print("GCD of the Array:", gcd_value)
```

# Q27) Find the Least Common Multiple (LCM) of an NumPy Array

```python
import numpy as np

arr = np.array([3, 4, 6])
lcm_value = np.lcm.reduce(arr)

print("LCM of the Array:", lcm_value)
```

# Q28) Find the Kth Smallest Element in a 1-D Array

```python
import numpy as np

arr = np.array([7, 10, 4, 3, 20, 15])
k = 3

sorted_arr = np.sort(arr)
kth_smallest = sorted_arr[k - 1]  # Kth smallest element

print(f"{k}th Smallest Element:", kth_smallest)
```

# Q29) Find the Kth Largest Element in a 1-D Array

```python
import numpy as np

arr = np.array([7, 10, 4, 3, 20, 15])
k = 2

sorted_arr = np.sort(arr)
kth_largest = sorted_arr[-k]  # Kth largest element

print(f"{k}th Largest Element:", kth_largest)
```

# Q30) Find Elements that are Greater than the Previous Element

```python
import numpy as np

arr = np.array([1, 2, 1, 4, 6, 3, 7])
greater_elements = arr[1:][arr[1:] > arr[:-1]]

print("Elements Greater than Previous:", greater_elements)
```
