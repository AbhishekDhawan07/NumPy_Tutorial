# Q1) Find the Maximum and Minimum Element in a 1-D  NumPy Array.


```python
import numpy as np

# Creating a 1D NumPy array

arr = np.array([12, 45, 2, 67, 34, 89, 10])

# Finding maximum and minimum elements

max_element = np.max(arr)
min_element = np.min(arr)
print("Maximum Element:", max_element)
print("Minimum Element:", min_element)
```

# Q2) Reverse a 1-D Array.



```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
reversed_arr = arr[::-1]  # Using slicing

print("Original Array:", arr)
print("Reversed Array:", reversed_arr)
```

# Q3) Find the Sum of Elements in a 1-D Array.


```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25])
sum_of_elements = np.sum(arr)

print("Sum of Elements:", sum_of_elements)
```

# Q4) Find the Mean, Median, and Standard Deviation of the NumPy 1- D Array.


```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

mean_value = np.mean(arr)
median_value = np.median(arr)
std_dev = np.std(arr)

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_dev)
```

# Q5) Remove Duplicates from a 1-D Array

```python
import numpy as np

arr = np.array([10, 20, 30, 20, 40, 10, 50])
unique_arr = np.unique(arr)  # Removes duplicate elements

print("Array with Unique Elements:", unique_arr)
```

# Q6) Sort a 1-D Array

```python
import numpy as np

arr = np.array([34, 1, 78, 23, 56, 90, 12])
sorted_arr = np.sort(arr)

print("Sorted Array:", sorted_arr)
```

# Q7) Find the Second Largest Element in 1-D Array 

```python
import numpy as np

arr = np.array([10, 20, 4, 45, 99])
sorted_arr = np.sort(arr)
second_largest = sorted_arr[-2]

print("Second Largest Element:", second_largest)
```

# Q8) Count Occurrences of an Element in 1- D Array.


```python
import numpy as np

arr = np.array([1, 2, 3, 2, 2, 3, 4, 5, 1])
element = 2
count = np.count_nonzero(arr == element)

print(f"Occurrence of {element}:", count)
```

# Q9) Find the Cumulative Sum of Elements in 1- D Array

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
cumulative_sum = np.cumsum(arr)

print("Cumulative Sum:", cumulative_sum)
```

# Q10) Find Missing Number in a Consecutive Series in 1- D Array

```python
import numpy as np

arr = np.array([1, 2, 3, 5, 6])
missing_number = set(range(arr[0], arr[-1] + 1)) - set(arr)

print("Missing Number:", missing_number)
```

# Q11) Compute the Dot Product of Two 1-D Arrays.

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
dot_product = np.dot(arr1, arr2)

print("Dot Product:", dot_product)
```

# Q12) Find the Intersection of Two 1- D Arrays

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([3, 4, 5, 6, 7])
intersection = np.intersect1d(arr1, arr2)

print("Intersection of Arrays:", intersection)
```

# Q13) Find the Union of Two 1- D Arrays

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
union_arr = np.union1d(arr1, arr2)

print("Union of Arrays:", union_arr)
```

# Q14) Find Indices of Non-Zero Elements in 1-D Array


```python
import numpy as np

arr = np.array([0, 1, 0, 2, 3, 0, 4])
nonzero_indices = np.nonzero(arr)

print("Indices of Non-Zero Elements:", nonzero_indices)
```

# Q15) Normalize a 1-D Array.


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
normalized_arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))

print("Normalized Array:", normalized_arr)
```
